A) Write a Python program for Handling Missing Value. Replace missing value ofsalary,
age column with mean of that column.(Use Data.csv file).

```python
dataset = pd.read_csv('Data.csv')
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset['Age'].fillna(dataset['Age'].mean(), inplace = True)
```

```python
dataset['Salary'].fillna(dataset['Salary'].mean(), inplace = True)
```

B) Write a Python program to create a Bar plot to get the frequency of the three species of the Iris
data. 

```python
dataset = pd.read_csv('iris.csv')
dataset
```

```python
dist = dataset['species'].value_counts()
plt.bar(dist.index, dist.values, color = 'skyblue')
plt.show()
```
