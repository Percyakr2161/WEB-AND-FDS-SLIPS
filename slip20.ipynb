```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

A) Generate a random array of 50 integers and display them using a line chart, scatter plot,
histogram and box plot. Apply appropriate color, labels and styling options.

```python
data = np.random.randint(1, 101, 50)

plt.subplot(2,2,1)
plt.plot(data, color = 'r')
plt.title("Line Plot")

plt.subplot(2,2,2)
plt.scatter(range(50), data, marker = "*")
plt.title("Scatter Plot")

plt.subplot(2,2,3)
plt.hist(data, color = "pink")
plt.title("HIstogram Plot")

plt.subplot(2,2,4)
sns.boxplot(data = data, color = 'g')
plt.title("Box Plot")

plt.tight_layout()
plt.show()
```

B) Add two outliers to the above data and display the box plot. 

```python
data_with_outliers = np.append(data, [150, -10])

plt.figure(figsize=(8, 6))
plt.boxplot(data_with_outliers, vert=False)
plt.title('Box Plot of Random Data with Outliers')
plt.xlabel('Values')
plt.show()
```
