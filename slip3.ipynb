```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("iris.csv")
dataset
```

```python
plt.subplot(2,2,1)
sns.boxplot(data = dataset, x = 'species', y = 'sepal_length', color = 'red')
plt.title("Species vs Sepal Length")

plt.subplot(2,2,2)
sns.boxplot(data = dataset, x = 'species', y = 'sepal_width', color = 'yellow')
plt.title("Species vs Sepal Width")

plt.subplot(2,2,3)
sns.boxplot(data = dataset, x = 'species', y = 'petal_length', color = 'orange')
plt.title("Species vs Petal Length")

plt.subplot(2,2,4)
sns.boxplot(data = dataset, x = 'species', y = 'petal_width', color = 'skyblue')
plt.title("Species vs Petal Width")

plt.tight_layout()
plt.show()
```
